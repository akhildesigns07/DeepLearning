```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.fashion_mnist.load_data()
X_train, X_test = X_train / 255.0, X_test / 255.0  # Normalize

# Flatten the images
X_train = X_train.reshape(-1, 28 * 28)
X_test = X_test.reshape(-1, 28 * 28)
```

```python
def create_deep_model(activation):
    model = keras.Sequential()
    model.add(layers.InputLayer(input_shape=(28 * 28,)))
    for _ in range(100):  # Add 100 hidden layers
        model.add(layers.Dense(64, activation=activation))
    model.add(layers.Dense(10, activation='softmax'))  # Output layer
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
activation_functions = ["sigmoid", "relu", "elu", "selu"]
histories = {}
```

```python
for activation in activation_functions:
    print(f"\nTraining model with {activation} activation:")
    model = create_deep_model(activation)
    history = model.fit(X_train, y_train, epochs=10, batch_size=32,
                        validation_split=0.1, verbose=1)
    test_loss, test_accuracy = model.evaluate(X_test, y_test)

    # Plot the loss and accuracy curves
    plt.plot(history.history['loss'], label=f"{activation} - Training Loss")
    plt.plot(history.history['val_loss'], label=f"{activation} - Validation Loss")
    plt.title(f"Loss Curves for {activation}")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()
```
